# Regression Metrics
Basic regression metrics used in machine learning are documented here

# 1. Mean Absolute Error - MAE
mae = 1/n * sum(|Ytrue - Ypred|)
* Less sensitive to outliers than MSE but doesnt penalize large errors as heavily
* MAE is less influenced by outliers but might give equal weight to small & large errors, which might not be desirable depending the applications.
## Use: your data contains outliers; Interpretability is key - same scale
## Avoid: large errors matter; when -ve & +ve errors have different meaning

# 2. Mean Square Error - MSE
mse = 1/n * sum( (Ytrue - Ypred)^2 )
* Lower is better, indicating closer predictions to the ground truth
* MSE peanlizes larger errors more heavily, making it suitable for regression problems where large errors are more critical.
## USE: when you have normally distributed errors; Penalizing large errors is crucial; Optimization algorithm work well with this.
## AVOID: Interpreability is needed; Outliers heavily influence your data; when same scale is needed

# 3. Root Mean Square Error - RMSE
rmse = sqrt( 1/n * sum( (Ytrue-Ypred)^2 ) )
* same Units
* easy to interpret
* But can be affected by outliers due to squaring errors.
## AVOID: when non-normal error distribution; Outliers heavily influence your data

# 4. R^2 Error
R2 = 1 - ( sum( [Ytrue-Ypred]^2) / sum( [Ytrue-Ymean]^2) )
* It represents the proportion of variance in the target variable explained bythe model. 
* Ranges from 0 to 1(perfect fit)
* R^2  of 0.8 indicates the model explains 80% of the target variables variance.
* Is easily interpretable but can be misleading for non-linear models or datasets with few features.
* It never decreaes no matter the no.of variables we add to our regression model.


# 5. Adjusted R^2
* Similar to  R^2 but adjusts for the no.of features, providing a more reliable estimate of the model fit.
* Adjusted R^2 of 0.75 is trustworthy than a regular R^2 of 0.8 with many features
